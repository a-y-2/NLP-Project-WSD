# TAGGING OF AMBIGUOUS WORDS

```python
import nltk
from nltk import word_tokenize, pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Function to perform POS tagging for each row in the DataFrame
def tag_ambiguous_word(row):
    prompt = row['prompts']
    ambiguous_word = row['ambiguous_words']

    # Tokenize the prompt sentence
    tokens = word_tokenize(prompt)

    # Perform POS tagging on the tokens
    tagged_tokens = pos_tag(tokens)

    # Find the POS tag for the ambiguous word
    ambiguous_word_tag = None
    for token, tag in tagged_tokens:
        if token == ambiguous_word:
            ambiguous_word_tag = tag
            break

    return ambiguous_word_tag

# Apply the function to each row of the DataFrame to tag the ambiguous words
df['ambiguous_word_tag'] = df.apply(tag_ambiguous_word, axis=1)

print(df)
```

# FIND SENSES OF AMBIGUOUS WORDS FROM WORDNET

```python
import pandas as pd
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
# Function to map NLTK POS tags to WordNet POS tags
def nltk_to_wordnet_pos(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None

# Function to find synsets (senses) of ambiguous words by their POS tag from WordNet
def find_word_senses(word, tag):
    wn_tag = nltk_to_wordnet_pos(tag)
    if wn_tag:
        synsets = wordnet.synsets(word, pos=wn_tag)
        return synsets
    else:
        return []

# Apply the function to the DataFrame to find the synsets for each ambiguous word
df['word_senses'] = df.apply(lambda row: find_word_senses(row['ambiguous_words'], row['ambiguous_word_tag']), axis=1)

print(df)
```

#  GET DEFINITION OF SENSES FROM WORDNET

```python
import pandas as pd
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
# Function to map NLTK POS tags to WordNet POS tags
def nltk_to_wordnet_pos(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None


def find_word_senses(word, tag):
    wn_tag = nltk_to_wordnet_pos(tag)
    if wn_tag:
        synsets = wordnet.synsets(word, pos=wn_tag)
        if synsets:
            definitions = [synset.definition() for synset in synsets]
            return definitions
        else:
            return None
    else:
        return None


# Apply the function to the DataFrame to find the synsets for each ambiguous word
df['word_senses'] = df.apply(lambda row: find_word_senses(row['ambiguous_words'], row['ambiguous_word_tag']), axis=1)

print(df)
```

## SEPARATE OUT THE SENSES INTO DIFFERENT ROWS

```python
import pandas as pd

def split_definitions(row):
    word_senses = row['word_senses']

    # Check if word_senses is a string
    if isinstance(word_senses, str):
        definitions = word_senses.split(',')
    else:
        # If it's not a string (e.g., it's a float), return an empty list
        definitions = []

    return pd.Series(definitions)

# Apply the function to each row and stack the resulting rows
expanded_df = df.apply(split_definitions, axis=1).stack().reset_index(level=1, drop=True).reset_index()
expanded_df.columns = ['original_index', 'word_sense']

# Merge with the original DataFrame to bring in other columns
expanded_df = expanded_df.merge(df.drop('word_senses', axis=1), how='left', left_on='original_index', right_index=True)

```

## GET THE CONTENT WORDS OF THE SENSES

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
# Function to filter content words from text
def filter_content_words(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text.lower())  # Tokenize and convert to lowercase
    filtered_tokens = [token for token in tokens if token.isalnum() and token not in stop_words]
    pos_tags = nltk.pos_tag(filtered_tokens)  # Get POS tags for filtered tokens
    content_words = [token for token, tag in pos_tags if tag.startswith(('N', 'V', 'J', 'R'))]
    return ' '.join(content_words)

# Apply the function to the 'word_sense' column
expanded_df['content_words_sense'] = expanded_df['word_sense'].apply(filter_content_words)
```

```python
expanded_df.head()
```

```python
# Apply the function to the 'prompts' column
expanded_df['content_words_prompts'] = expanded_df['prompts'].apply(filter_content_words)
```

```python
expanded_df.head()
```

```python
# Function to remove words from content_words_sense column
def remove_ambiguous_words(row):
    sense_words = row['content_words_prompts'].split()
    ambiguous_words = row['ambiguous_words'].split()
    filtered_words = [word for word in sense_words if word not in ambiguous_words]
    return ' '.join(filtered_words)

# Apply the function to remove ambiguous words
expanded_df['content_words_prompts'] = expanded_df.apply(remove_ambiguous_words, axis=1)
```

```python
# Function to create pairs of words
def create_word_pairs(row):
    sense_words = row['content_words_sense'].split()
    prompt_words = row['content_words_prompts'].split()
    pairs = [(sense, prompt) for sense in sense_words for prompt in prompt_words]
    return pairs

# Apply the function to create pairs of words
expanded_df['word_pairs'] = expanded_df.apply(create_word_pairs, axis=1)
```

```python
# Separate each pair in 'filtered_word_pairs' into different rows
new_rows = []
for index, row in expanded_df.iterrows():
    original_index = row['original_index']
    word_sense = row['word_sense']
    prompts = row['prompts']
    ambiguous_words = row['ambiguous_words']
    tagged_sentences = row['tagged_sentences']
    ambiguous_word_tag = row['ambiguous_word_tag']
    content_words_sense = row['content_words_sense']
    content_words_prompts = row['content_words_prompts']
    word_pairs = row['word_pairs']

    for pair in word_pairs:
        # Combine the word pair into a single string
        word_pair_str = f"({pair[0]}, {pair[1]})"

        new_rows.append({
            'original_index': original_index,
            'word_sense': word_sense,
            'prompts': prompts,
            'ambiguous_words': ambiguous_words,
            'tagged_sentences': tagged_sentences,
            'ambiguous_word_tag': ambiguous_word_tag,
            'content_words_sense': content_words_sense,
            'content_words_prompts': content_words_prompts,
            'word_pair': word_pair_str
        })

# Create a new DataFrame from the list of new rows
new_df = pd.DataFrame(new_rows)

# Display the new DataFrame
print(new_df)
```

```python
import numpy as np 
def calculate_formula(x,w1,w2):
    try:
        result = np.log((x * 3544633) / (w1 * w2 * 2187)) / np.log(2)
    except ZeroDivisionError:
        result = 0
    return result
```

```python
final_df_nlp['collocation_score'] = final_df_nlp.apply(lambda row: calculate_formula(row['freq'],row['w1'],row['w2']), axis=1)
```

```python

# Replace -inf values with NaN for handling
final_df_nlp['collocation_score'].replace(-np.inf, np.nan, inplace=True)

# Group by 'content_words_sense' and find the maximum 'collocation_score' within each group
max_scores = final_df_nlp.groupby('content_words_sense')['collocation_score'].max().reset_index()

# Merge the max_scores DataFrame back to the original DataFrame df
df_merged = pd.merge(final_df_nlp, max_scores, on='content_words_sense', suffixes=('', '_max'))

# Rename the 'collocation_score_max' column to 'max_score'
df_merged.rename(columns={'collocation_score_max': 'max_score'}, inplace=True)

# Find the corresponding sense for each maximum score
max_sense = df_merged.loc[df_merged['collocation_score'] == df_merged['max_score'], ['content_words_sense', 'max_score']]

# Merge max_sense back to df_merged to append the corresponding sense as a new column
df_final = pd.merge(df_merged, max_sense, on='max_score', suffixes=('', '_sense'))

# Drop unnecessary columns and rename the final DataFrame
df_final.drop('max_score', axis=1, inplace=True)
df_final.rename(columns={'content_words_sense_sense': 'sense_with_max_score'}, inplace=True)
```
